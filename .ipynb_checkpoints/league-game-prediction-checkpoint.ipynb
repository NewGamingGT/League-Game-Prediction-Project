# League Game Prediction Project

This project is trying to predict the out come of the game by using the league data sets found here: https://www.kaggle.com/datasnaek/league-of-legends.

**Note: This data set is last updated about 2 years ago (2018), hence it is a bit outdated. You will find missing new champions as it has not been released at that time.**

## Features

**Data dictionary:**
* Game ID
* Creation Time (in Epoch format)
* Game Duration (in seconds)
* Season ID
* Winner (1 = team1, 2 = team2) (team 1 = blue team (bottom left)) (team 2 = red team (top right))
* First Baron, dragon, tower, blood, inhibitor and Rift Herald (1 = team1, 2 = team2, 0 = none)
* Champions and summoner spells for each team (Stored as Riot's champion and summoner spell IDs)
* The number of tower, inhibitor, Baron, dragon and Rift Herald kills each team has
* The 5 bans of each team (Again, champion IDs are used)

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.svm import LinearSVC, SVC
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score, plot_roc_curve

import pickle
```

```python
league_dataset = pd.read_csv("games.csv")
league_dataset.head()
```

## Finding the correlation

```python
# Droping non useful data that will not affect the outcome of the game significantly
corr_matrix = league_dataset.drop(["gameId", "creationTime", "seasonId", "t1_champ1_sum1", "t1_champ1_sum2", 
                     "t1_champ2_sum1", "t1_champ2_sum2", "t1_champ3_sum1", "t1_champ3_sum2",
                     "t1_champ4_sum1", "t1_champ4_sum2", "t1_champ5_sum1", "t1_champ5_sum2",
                     "t2_champ1_sum1", "t2_champ1_sum2", "t2_champ2_sum1", "t2_champ2_sum2", 
                     "t2_champ3_sum1", "t2_champ3_sum2", "t2_champ4_sum1", "t2_champ4_sum2", 
                     "t2_champ5_sum1", "t2_champ5_sum2", "t1_ban1", "t1_ban2", "t1_ban3",
                     "t1_ban4", "t1_ban5", "t2_ban1", "t2_ban2", "t2_ban3",
                     "t2_ban4", "t2_ban5", "t1_champ1id", "t1_champ2id", "t1_champ3id",
                     "t1_champ4id", "t1_champ5id", "t2_champ1id", "t2_champ2id", "t2_champ3id",
                     "t2_champ4id", "t2_champ5id"], axis=1).corr()
fig, ax = plt.subplots(figsize=(15, 10))
ax = sns.heatmap(corr_matrix, annot=True, linewidths=0.5, fmt=".2f", cmap="YlGnBu")
```

## Plotting the data

Finding the relationships between the first objective taken vs the number of games won

```python
ind = np.arange(3)  # the x locations for the groups
width = 0.3       # the width of the bars

fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(13, 20))
sns.set_style("whitegrid")

first_objective_taken = [[pd.crosstab(league_dataset.winner, league_dataset.firstTower), 
                                        "First Tower taken vs Number of games won",
                                        "First Tower taken by"],
                         [pd.crosstab(league_dataset.winner, league_dataset.firstBlood),
                                        "First Blood taken vs Number of games won",
                                        "First Blood taken by"],
                         [pd.crosstab(league_dataset.winner, league_dataset.firstInhibitor),
                                            "First Inhibitor taken vs Number if games won",
                                            "First Inhibitor taken by"],
                         [pd.crosstab(league_dataset.winner, league_dataset.firstDragon),
                                         "First Dragon taken vs Number of games won",
                                         "First Dragon taken by"],
                         [pd.crosstab(league_dataset.winner, league_dataset.firstBaron),
                                        "First Baron taken vs Number of games won",
                                        "First Baron taken by"],
                         [pd.crosstab(league_dataset.winner, league_dataset.firstRiftHerald),
                                            "First Rift Herald taken vs Number of games won",
                                            "First Rift Herald taken by"]]

def draw_bar(row, column, first_objective_taken):
    ax[row][column].bar(x=ind, width=width, height=first_objective_taken[0].loc[1], color="lightblue")
    ax[row][column].bar(x=ind+width, width=width, height=first_objective_taken[0].loc[2], color="salmon")

    ax[row][column].set_ylabel('Number of games won')
    ax[row][column].set_title(first_objective_taken[1])
    ax[row][column].set_xticks(ind + width / 2)
    ax[row][column].set_xticklabels(("None of the teams", "Blue Team", "Red Team"))
    ax[row][column].set_xlabel(first_objective_taken[2])
    ax[row][column].legend(labels=["Blue Team", "Red Team"], title="Winners")

draw_bar(row=0, column=0, first_objective_taken=first_objective_taken[0])
draw_bar(row=0, column=1, first_objective_taken=first_objective_taken[1])

draw_bar(row=1, column=0, first_objective_taken=first_objective_taken[2])
draw_bar(row=1, column=1, first_objective_taken=first_objective_taken[3])

draw_bar(row=2, column=0, first_objective_taken=first_objective_taken[4])
draw_bar(row=2, column=1, first_objective_taken=first_objective_taken[5])
```

### Conclusion
Which ever team gets the first objective, their probability of winning the game is much higher. Taking the first inhibitor will significantly increase the team chance of winning

## Modelling the data

Create a machine learning model

```python
# Dropping not useful column that will not affect the outcome of the game significantly
df = league_dataset.drop(["gameId", "creationTime", "seasonId", "t1_champ1_sum1", "t1_champ1_sum2", 
                     "t1_champ2_sum1", "t1_champ2_sum2", "t1_champ3_sum1", "t1_champ3_sum2",
                     "t1_champ4_sum1", "t1_champ4_sum2", "t1_champ5_sum1", "t1_champ5_sum2",
                     "t2_champ1_sum1", "t2_champ1_sum2", "t2_champ2_sum1", "t2_champ2_sum2", 
                     "t2_champ3_sum1", "t2_champ3_sum2", "t2_champ4_sum1", "t2_champ4_sum2", 
                     "t2_champ5_sum1", "t2_champ5_sum2", "t1_ban1", "t1_ban2", "t1_ban3",
                     "t1_ban4", "t1_ban5", "t2_ban1", "t2_ban2", "t2_ban3",
                     "t2_ban4", "t2_ban5", "t1_champ1id", "t1_champ2id", "t1_champ3id",
                     "t1_champ4id", "t1_champ5id", "t2_champ1id", "t2_champ2id", "t2_champ3id",
                     "t2_champ4id", "t2_champ5id", "gameDuration", "t1_riftHeraldKills", "t2_riftHeraldKills"], axis=1)
df.head()
```

```python
np.random.seed(42)

x = df.drop("winner", axis=1)
y = df.winner

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
models = {"Linear SVC": LinearSVC(),
          "Random Forest": RandomForestClassifier(),
          "Logistic Regression": LogisticRegression(),
          "SGDClassifier": SGDClassifier(),
          "SVC": SVC(),
          "KNN": KNeighborsClassifier()}

model_scores = {}

for model_name, model in models.items():
    model.fit(x_train, y_train)
    model_scores[model_name] = model.score(x_test, y_test)
    
model_scores
```

```python
model_score_df = pd.DataFrame(model_scores, index=["accuracy"])
sns.set_style("whitegrid")
ax = model_score_df.T.plot(kind="bar", ylim=0.9, rot=0, figsize=(10, 6), legend=False, color="lightblue")
ax.set_ylabel("Accuracy");
```

## Eliminating the models
Eliminated SGD Classifier due to it's low accuracy compared to the other machine learning models

## Testing Linear SVC

```python
# Tuning Linear SVC
np.random.seed(42)

lin_svc_grid = {"C": np.linspace(0, 1000, 50),
               "max_iter": [100, 1000, 5000, 10000]}

rs_lin_svc = RandomizedSearchCV(LinearSVC(), param_distributions=lin_svc_grid, cv=5, n_iter=5, verbose=True)
rs_lin_svc.fit(x_train, y_train)
```

## Evaluating the model

```python
model = LogisticRegression(C=0.07906043210907697, penalty="l1", solver="saga").fit(x_test, y_test).fit(x_test, y_test)
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
y_test
```

```python
plot_roc_curve(model, x_test, y_test)
```

```python
sns.set(font_scale=1.5)

fig, ax = plt.subplots(figsize=(3, 3))
ax = sns.heatmap(confusion_matrix(y_test, y_preds),
                 annot=True,
                 cbar=False,
                 fmt="g")
plt.xlabel("True label")
plt.ylabel("Predicted label")
```

```python
print(classification_report(y_test, y_preds))
```

```python
scorings = {"accuracy": None,
            "precision": None, 
            "recall": None, 
            "f1": None}

for scoring in scorings.keys():
    scorings[scoring] = np.mean(cross_val_score(model, x, y, cv=5, scoring=scoring))

scorings
```

```python
cv_metrics = pd.DataFrame(scorings, index=[0])
ax = cv_metrics.T.plot.bar(title="Cross-validated classification metrics",
                      legend=False, ylim=0.9);
ax.set_ylabel("Accuracy")
```
